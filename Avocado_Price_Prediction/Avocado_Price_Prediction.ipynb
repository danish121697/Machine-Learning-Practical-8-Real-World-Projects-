```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import random
import seaborn as sns
from fbprophet import Prophet
```

```python
avo_df = pd.read_csv('avocado.csv')
```

```python
#Will predict prices using price column only
#Ignore all other columns
avo_df
```

```python
avo_df.head()
```

```python
avo_df.tail()
```

```python
avo_df = avo_df.sort_values("Date")
```

```python
avo_df
```

```python
#Price moves varies with different values
plt.figure(figsize = (10,10))
plt.plot(avo_df.Date,avo_df.AveragePrice)
```

```python
#Data balanced across regions
plt.figure(figsize = [25,25])
sns.countplot(x = 'region',data = avo_df)
plt.xticks(rotation = 45)
```

```python
plt.title("Number of avocados per Year")
sns.countplot(x = 'year',data = avo_df)
#2018 lower as year wasn't complete
```

```python
avo_df = avo_df[['Date','AveragePrice']]
```

```python
avo_df
```

```python
avo_df =  avo_df.rename(columns = {'Date':'ds','AveragePrice':'y'})
```

```python
avo_df
```

```python
#Train model
m = Prophet()
m.fit(avo_df)
```

```python
#forecast the future
future = m.make_future_dataframe(periods = 365)
forecast = m.predict(future)
```

```python
#Trends given
#Datas in the future such as 2019 given
#Future forecasts appended to dataframe
forecast
```

```python
figure = m.plot(forecast, xlabel = 'Date', ylabel = 'Price')
```

```python
#Predicting Trend
figure3 = m.plot_components(forecast)
```

```python
avocado_df = pd.read_csv('avocado.csv')
```

```python
avocado_df_sample = avocado_df[avocado_df['region']=='West']
avocado_df_sample
```

```python
avocado_df_sample = avocado_df_sample.sort_values("Date")
avocado_df_sample
```

```python
plt.figure(figsize=(10,10))
plt.plot(avocado_df_sample['Date'], avocado_df_sample['AveragePrice'])
```

```python
avocado_df_sample = avocado_df_sample.rename(columns={'Date':'ds', 'AveragePrice':'y'})
```

```python
m = Prophet()
m.fit(avocado_df_sample)
# Forcasting into the future
future = m.make_future_dataframe(periods=365)
forecast = m.predict(future)
```

```python
figure = m.plot(forecast, xlabel='Date', ylabel='Price')
```

```python
#Trend differs as before shown for the entire market
#Instead there is an increase
figure3 = m.plot_components(forecast)
```
